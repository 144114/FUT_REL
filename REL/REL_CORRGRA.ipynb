```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.DataFrame({'Crude Oil': CL['Close'], 'MNQ': MNQ['Close']}).dropna()
rolling_correlation = data['Crude Oil'].rolling(window=1 * 288).corr(data['MNQ'])

plt.figure(figsize=(14, 6))

plt.fill_between(rolling_correlation.index, rolling_correlation, 0, where=(rolling_correlation > 0), color='green', alpha=0.3)
plt.fill_between(rolling_correlation.index, rolling_correlation, 0, where=(rolling_correlation < 0), color='red', alpha=0.3)

plt.title("Rolling Correlation between Crude Oil and MNQ (Nasdaq-100 Micro E-mini Futures)")
plt.xlabel("Date")
plt.ylabel("Correlation")
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.DataFrame({'Crude Oil': CL['Close'], 'MNQ': MNQ['Close']}).dropna()
data_normalized = (data - data.min()) / (data.max() - data.min())

correlation = data['Crude Oil'].corr(data['MNQ'])
print("Correlation between Crude Oil and MNQ (Nasdaq-100 Micro E-mini Futures):", correlation)

plt.figure(figsize=(14, 6))
plt.plot(data_normalized['Crude Oil'], label='Crude Oil (Normalized)', color='black')
plt.plot(data_normalized['MNQ'], label='MNQ (Normalized)', color='red')
plt.title(f"Crude Oil vs. MNQ - Correlation: {correlation:.2f}")
plt.xlabel("Date")
plt.ylabel("Normalized Price")
plt.legend()
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.show()
```
